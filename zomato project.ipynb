 ## Zomato data analysis project 

### Step-1

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step-2 Create the data frame 

```python
dataframe=pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

### Step 3-Convert the datatype of col rate

```python
def handleRate(value):
    value=str(value).split("/") #it convert value into string and sptlit it from / for ex-4.1/5 so it silt  4.1 and /5
    value=value[0]              #it return 0th index means like above example in 0th index 4.1 so it return 4.1
    return float(value)         #it return final value as float

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

###                                                     Type of resturants

```python
sns.countplot(x=dataframe['listed_in(type)'], palette='viridis', hue=dataframe['listed_in(type)'])
plt.xlabel('Type of resturant')
```

#### concusion- Majority of the resturant falls in dining category

## Votes of resturant

```python
dataframe.head()
```

```python
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="green",marker='o')
plt.xlabel("Type of resturant",c="red",size=20)
plt.ylabel("Votes",c="red",size=20)
plt.show()
```

#### concusion- Dining resturant has recived maximum votes 

## Rating that the majority of restaurants have received 

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("rating distribution")
plt.show()
```

#### Conclusion-The majority resturants recived ratings from 3.5 to 4.0

# Average order spending by couples 

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data,palette='viridis',hue=couple_data)
plt.show()
```

#### Conclusion-The majority of couples preferr resturants with an approximate cost 300 rupees

# Which mode recives maximum rating 

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=dataframe,palette='Set2',hue='online_order')
plt.show()
```

#### concusion-Offline order recive lower rating than online order

# Which type of resturant rescive more offline order

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table, annot=True,cmap='coolwarm', fmt='d')
plt.title("Heat Map")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

#### conclusion-Dining restaurants primarily accept offlinr orders where as cafes primarily recives online orders. This suggests that clints prefer to place orders in person at restaurants,but prefer online ordering att cafes
